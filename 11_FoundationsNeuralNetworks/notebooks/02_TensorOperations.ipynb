## 1. Transposition

Transposition reverses the order of tensor indices.

- **Rank 1 (vector):** Transposition has no visible effect (it remains a 1‑D array).
- **Rank 2 (matrix):** Swaps rows and columns: `(A^T)_{i,j} = A_{j,i}`.
- **Rank 3 and above:** Reverses all indices: `(T^T)_{i,j,k} = T_{k,j,i}`.

**Note:** In PyTorch, using `.T` on tensors of dimension other than 2 is deprecated. Use `.mT` for batched matrices or `.permute()` for general cases.

```python
# Rank 1 vector – transposition has no effect
x = torch.tensor([1, 2, 3])
print(f"Vector: {x}")
print(f"Transposed vector: {x.T} (same)")
print("-" * 40)

# Rank 2 matrix – swaps rows and columns
A = torch.tensor([
    [0, 5, 1],
    [3, 0, 2],
    [0, 2, 0],
    [5, 3, 1]
])
print(f"Matrix A:\n{A}")
print(f"Transposed A (A.T):\n{A.T}")
print(f"Shape: {A.shape} -> {A.T.shape}")
print("-" * 40)

# Rank 3 tensor – reverses indices
T = torch.tensor([
    [[1, 2, 3], [4, 5, 6], [1, 2, 3], [1, 0.5, 2]],
    [[7, 8, 9], [10, 11, 12], [20, 1, 0], [0, 0, 1]]
])
print(f"Tensor T shape: {T.shape}")
print(f"Transposed T shape: {T.T.shape}")
print(f"T.T:\n{T.T}")
```

## 2. Permutation

While transposition reverses all dimensions, `transpose()` allows swapping any two specific dimensions.

```python
print(f"Original T shape: {T.shape}")

# Swap dimension 0 and 1
print(f"T.transpose(0,1) shape: {T.transpose(0, 1).shape}")
print(f"T.transpose(0,1):\n{T.transpose(0, 1)}")

# Swap dimension 1 and 2
print(f"\nT.transpose(1,2) shape: {T.transpose(1, 2).shape}")
print(f"T.transpose(1,2):\n{T.transpose(1, 2)}")
```

## 3. Tensor Addition

Addition is performed element‑wise. Both tensors must have the same shape.

```python
# Vector addition
x = torch.tensor([1, 2, 1])
y = torch.tensor([1, 2, 2])
print(f"{x} + {y} = {x + y}")
print("-" * 40)

# Matrix addition
A = torch.tensor([[0, 5, 1], [3, 0, 2], [0, 2, 0], [5, 3, 1]])
B = torch.tensor([[10, 0, 2], [9, 4, 1], [8, 1, 0], [1, 0, 0]])
print(f"A + B =\n{A + B}")
print("-" * 40)

# Higher‑rank tensor addition
T1 = torch.rand(2, 3, 4)
T2 = torch.rand(2, 3, 4)
print(f"T1 + T2 =\n{T1 + T2}")
```

## 4. Scalar Multiplication

Multiplying a tensor by a scalar scales every element: `(λT)_{i} = λ * T_{i}`.

```python
scalar = 2.5
print(f"{scalar} * {x} = {scalar * x}")
print(f"{scalar} * A =\n{scalar * A}")
print(f"{scalar} * T =\n{scalar * T}")
```

## 5. Dot Product (Inner Product)

The dot product of two vectors is: `x · y = Σᵢ xᵢ * yᵢ`. In PyTorch, use `torch.matmul()` or `torch.dot()`.

```python
dot_product = torch.matmul(x, y)
print(f"{x} · {y} = {dot_product}")
print(f"Alternative: torch.dot(x, y) = {torch.dot(x, y)}")
```

## 6. Matrix‑Vector and Matrix‑Matrix Multiplication

- **Matrix‑Vector:** `(A x)ᵢ = Σⱼ Aᵢⱼ xⱼ`
- **Matrix‑Matrix:** `(A B)ᵢⱼ = Σₖ Aᵢₖ Bₖⱼ`

Both are performed with `torch.matmul()`.

```python
product = torch.matmul(A, x)
print(f"A = \n{A}")
print(f"x = {x}")
print(f"A · x = {product}")
```

```python
rank2_product = torch.matmul(A, B.T)
print(f"A · B.T =\n{rank2_product}")
```

## 7. Multiplication with Higher‑Rank Tensors

Tensors of rank 3 or higher can be multiplied with vectors, matrices, or other tensors, provided the dimensions align.

- **Tensor‑Vector:** `(T x)ᵢⱼ = Σₖ Tᵢⱼₖ xₖ`
- **Tensor‑Matrix:** `(T A)ᵢⱼₖ = Σₗ Tᵢⱼₗ Aₗₖ`
- **Tensor‑Tensor:** `(T U)ᵢⱼₖ = Σₗ Tᵢⱼₗ Uₗₖ`

```python
# Create tensors
T3 = torch.rand(3, 3, 3)
A3 = torch.rand(3, 2)
x3 = torch.rand(3)

# Tensor‑Vector
print("Tensor (3x3x3) · Vector (3):")
print(f"T3:\n{T3}")
print(f"x3: {x3}")
print(f"Result:\n{torch.matmul(T3, x3)}")
print("-" * 40)

# Tensor‑Matrix
print("Tensor (3x3x3) · Matrix (3x2):")
print(f"Result:\n{torch.matmul(T3, A3)}")
print("-" * 40)

# Tensor‑Tensor
T4 = torch.rand(3, 3, 3)
print("Tensor (3x3x3) · Tensor (3x3x3):")
print(f"Result:\n{torch.matmul(T3, T4)}")
```

## 8. Hadamard (Element‑Wise) Product

The Hadamard product multiplies corresponding elements of two tensors of the same shape: `(T ⊙ U)ᵢ = Tᵢ * Uᵢ`. In PyTorch, use the `*` operator.

```python
hadamard = x * y
print(f"{x} ⊙ {y} = {hadamard}")
print(f"Matrix Hadamard product:\n{A * B}")
```

```python
outer_product = torch.outer(x, y)
print(f"{x} ⊗ {y} = \n{outer_product}")
```
