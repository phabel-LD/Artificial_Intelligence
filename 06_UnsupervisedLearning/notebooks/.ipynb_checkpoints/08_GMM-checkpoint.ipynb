## 1. Theoretical Background

Gaussian Mixture Models (GMM) are a probabilistic extension of K‑Means clustering. Instead of assigning each point to a single cluster (hard assignment), GMM assigns a probability (soft assignment) that a point belongs to each cluster.

**Key Concepts:**
- **Mixture Model:** Data is assumed to be generated from a weighted sum of K Gaussian distributions.
- **Expectation‑Maximisation (EM) Algorithm:** Iteratively estimates the parameters (means, covariances, mixing weights) of the Gaussians.
  - **E‑Step:** Compute the probability (responsibility) of each point belonging to each component.
  - **M‑Step:** Update the parameters to maximise the likelihood given the responsibilities.
- **Covariance Types:** `full`, `tied`, `diag`, `spherical` – control the shape and orientation of the Gaussians.
- **Model Selection:** Use AIC (Akaike Information Criterion) or BIC (Bayesian Information Criterion) to select the optimal number of components.

**GMM vs K‑Means:**
| Aspect | K‑Means | GMM |
|--------|---------|-----|
| Assignment | Hard (one cluster per point) | Soft (probability per cluster) |
| Cluster Shape | Spherical | Ellipsoidal (with covariance) |
| Sensitivity to Scale | Yes | Yes (but handles covariance) |
| Complexity | O(n * k * d) | O(n * k * d^2) |

## 2. Synthetic Data Experiment: Height and Weight

We generate synthetic height‑weight data and compare K‑Means and GMM clustering results.

```python
# %% Generate Synthetic Height-Weight Data
np.random.seed(42)

# Generate data for 4 clusters with different distributions
n_samples = 200

# Cluster 1: Short, light
height1 = np.random.normal(165, 5, n_samples // 4)
weight1 = np.random.normal(60, 5, n_samples // 4)

# Cluster 2: Tall, heavy
height2 = np.random.normal(185, 5, n_samples // 4)
weight2 = np.random.normal(85, 6, n_samples // 4)

# Cluster 3: Short, heavy
height3 = np.random.normal(168, 4, n_samples // 4)
weight3 = np.random.normal(78, 5, n_samples // 4)

# Cluster 4: Tall, light
height4 = np.random.normal(180, 4, n_samples // 4)
weight4 = np.random.normal(65, 4, n_samples // 4)

# Combine data
heights = np.concatenate([height1, height2, height3, height4])
weights = np.concatenate([weight1, weight2, weight3, weight4])

df_synth = pd.DataFrame({'Weight': weights, 'Height': heights})

print(f"Dataset shape: {df_synth.shape}")
print("\nFirst 5 rows:")
print(df_synth.head())
print("\nSummary statistics:")
print(df_synth.describe())

# Visualise the raw data
plt.figure(figsize=(8, 6))
plt.scatter(df_synth['Weight'], df_synth['Height'], alpha=0.6, color='steelblue')
plt.title('Synthetic Height‑Weight Data')
plt.xlabel('Weight (kg)')
plt.ylabel('Height (cm)')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('plots/synthetic_data.png', dpi=300)
plt.show()
```

### 2.1 K‑Means Clustering (Hard Assignment)

We first apply K‑Means to establish a baseline.

```python
# %% K-Means Clustering
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
kmeans_labels = kmeans.fit_predict(df_synth)

# Visualise K-Means clusters
plt.figure(figsize=(8, 6))
colors = ['blue', 'green', 'orange', 'purple']
for k in range(4):
    cluster_data = df_synth[kmeans_labels == k]
    plt.scatter(cluster_data['Weight'], cluster_data['Height'],
                c=colors[k], alpha=0.6, label=f'Cluster {k+1}')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            s=200, c='red', marker='X', label='Centroids')
plt.title('K-Means Clustering (K=4)')
plt.xlabel('Weight (kg)')
plt.ylabel('Height (cm)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('plots/kmeans_clusters.png', dpi=300)
plt.show()
```

### 2.2 GMM Clustering (Soft Assignment)

Now we apply GMM, which provides soft assignments (probabilities) for each point.

```python
# %% GMM Clustering
gmm = GaussianMixture(n_components=4, random_state=42)
gmm_labels = gmm.fit_predict(df_synth)
gmm_proba = gmm.predict_proba(df_synth)  # Soft assignments

# Visualise GMM clusters (hard assignment for visualisation)
plt.figure(figsize=(8, 6))
for k in range(4):
    cluster_data = df_synth[gmm_labels == k]
    plt.scatter(cluster_data['Weight'], cluster_data['Height'],
                c=colors[k], alpha=0.6, label=f'Cluster {k+1}')
# Plot means (centroids)
plt.scatter(gmm.means_[:, 0], gmm.means_[:, 1],
            s=200, c='red', marker='X', label='Means')
plt.title('GMM Clustering (K=4)')
plt.xlabel('Weight (kg)')
plt.ylabel('Height (cm)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('plots/gmm_clusters.png', dpi=300)
plt.show()

# Compare silhouette scores
from sklearn.metrics import silhouette_score
sil_kmeans = silhouette_score(df_synth, kmeans_labels)
sil_gmm = silhouette_score(df_synth, gmm_labels)
print(f"Silhouette Score (K-Means): {sil_kmeans:.4f}")
print(f"Silhouette Score (GMM):      {sil_gmm:.4f}")
```

### 2.3 Visualising Soft Assignments

GMM assigns a probability to each cluster for every point. We visualise the uncertainty for a sample point.

```python
# %% Visualise Soft Assignments
# Pick a sample point with high uncertainty
idx = np.argmax(gmm_proba.std(axis=1))  # highest standard deviation across cluster probabilities
point = df_synth.iloc[idx]
print(f"Sample point: Weight={point['Weight']:.2f}, Height={point['Height']:.2f}")
print(f"Cluster probabilities: {gmm_proba[idx].round(3)}")

# Plot the point with its probabilities
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Scatter plot with the point highlighted
ax1.scatter(df_synth['Weight'], df_synth['Height'], c=gmm_labels, cmap='viridis', alpha=0.5)
ax1.scatter(point['Weight'], point['Height'], s=200, c='red', marker='*', label='Selected Point')
ax1.set_title('GMM Clustering with Highlighted Point')
ax1.set_xlabel('Weight (kg)')
ax1.set_ylabel('Height (cm)')
ax1.legend()
ax1.grid(True, alpha=0.3)

# Bar plot of probabilities
ax2.bar(range(1, 5), gmm_proba[idx], color=['blue', 'green', 'orange', 'purple'])
ax2.set_title('Cluster Probabilities for Selected Point')
ax2.set_xlabel('Cluster')
ax2.set_ylabel('Probability')
ax2.set_ylim(0, 1)
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('plots/gmm_soft_assignments.png', dpi=300)
plt.show()
```

## 3. Model Selection: AIC and BIC

We use AIC and BIC to determine the optimal number of components for the GMM.

```python
# %% Model Selection with AIC/BIC
n_components_range = range(1, 10)
aic_scores = []
bic_scores = []

for n_comp in n_components_range:
    gmm_temp = GaussianMixture(n_components=n_comp, random_state=42)
    gmm_temp.fit(df_synth)
    aic_scores.append(gmm_temp.aic(df_synth))
    bic_scores.append(gmm_temp.bic(df_synth))

plt.figure(figsize=(10, 6))
plt.plot(n_components_range, aic_scores, marker='o', label='AIC', linewidth=2)
plt.plot(n_components_range, bic_scores, marker='s', label='BIC', linewidth=2)
plt.title('AIC and BIC for GMM Model Selection')
plt.xlabel('Number of Components')
plt.ylabel('Score (lower is better)')
plt.xticks(n_components_range)
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('plots/gmm_aic_bic.png', dpi=300)
plt.show()

# Optimal number of components
optimal_aic = n_components_range[np.argmin(aic_scores)]
optimal_bic = n_components_range[np.argmin(bic_scores)]
print(f"Optimal components (AIC): {optimal_aic}")
print(f"Optimal components (BIC): {optimal_bic}")
print(f"True number of components: 4")
```

## 4. Real‑World Case Study: Market Regime Detection

We apply GMM to financial market data to identify different market regimes based on the relationship between CPI inflation and S&P 500 returns.

**Data:** Monthly CPI inflation (synthetic) and S&P 500 (SPY) returns from 2000 to 2024.

**Hypothesis:** Different economic environments (e.g., high inflation, low inflation, recession) correspond to different clusters in the inflation‑return space.

```python
# %% Load Financial Data
print("Downloading S&P 500 (SPY) data from Yahoo Finance...")
try:
    spy_data = yf.download('SPY', start='2000-01-01', end='2024-09-01', progress=False)
    spy_monthly_returns = spy_data['Close'].resample('ME').ffill().pct_change().dropna()
    print("Data downloaded successfully.")
except Exception as e:
    print(f"Download failed: {e}")
    # Generate synthetic market data
    print("Generating synthetic market data...")
    np.random.seed(42)
    dates = pd.date_range('2000-01-01', periods=300, freq='ME')
    spy_monthly_returns = pd.Series(np.random.normal(0.005, 0.04, len(dates)), index=dates)

# Generate synthetic CPI data (using normal distribution, mean 2% annual inflation)
np.random.seed(42)
dates = spy_monthly_returns.index
cpi_data = np.random.normal(loc=0.02 / 12, scale=0.01 / 12, size=len(dates))  # Monthly inflation
cpi_monthly = pd.Series(cpi_data, index=dates)

# Combine data
combined_data = pd.concat([cpi_monthly, spy_monthly_returns], axis=1).dropna()
combined_data.columns = ['CPI Inflation', 'Market Return']

print(f"\nData shape: {combined_data.shape}")
print("\nFirst 5 rows:")
print(combined_data.head())
print("\nSummary statistics:")
print(combined_data.describe())
```

```python
# %% GMM on Financial Data
gmm_fin = GaussianMixture(n_components=3, random_state=42)
gmm_fin.fit(combined_data)
fin_labels = gmm_fin.predict(combined_data)

print(f"GMM converged in {gmm_fin.n_iter_} iterations.")
print(f"Cluster weights: {gmm_fin.weights_.round(3)}")

# Visualise clusters
plt.figure(figsize=(10, 7))
scatter = plt.scatter(combined_data['CPI Inflation'], combined_data['Market Return'],
                      c=fin_labels, cmap='viridis', alpha=0.6, edgecolors='k', linewidth=0.3)
plt.scatter(gmm_fin.means_[:, 0], gmm_fin.means_[:, 1],
            s=300, c='red', marker='X', label='Cluster Means')
plt.xlabel('CPI Inflation (Monthly)')
plt.ylabel('Market Return (Monthly)')
plt.title('GMM Clustering: Inflation vs Market Returns')
plt.colorbar(scatter, label='Cluster')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('plots/gmm_financial_clusters.png', dpi=300)
plt.show()

# Interpret the clusters
cluster_profiles = pd.DataFrame({
    'Cluster': range(3),
    'Mean Inflation': gmm_fin.means_[:, 0],
    'Mean Return': gmm_fin.means_[:, 1],
    'Weight': gmm_fin.weights_
})
print("\nCluster Profiles:")
print(cluster_profiles)
```

### 4.1 Density Contours

We plot the GMM density contours to visualise the fitted mixture components.

```python
# %% GMM Density Contours
x_min, x_max = combined_data['CPI Inflation'].min() - 0.005, combined_data['CPI Inflation'].max() + 0.005
y_min, y_max = combined_data['Market Return'].min() - 0.05, combined_data['Market Return'].max() + 0.05

xx, yy = np.meshgrid(np.linspace(x_min, x_max, 200),
                     np.linspace(y_min, y_max, 200))

XY = np.column_stack([xx.ravel(), yy.ravel()])
Z = -gmm_fin.score_samples(XY)
Z = Z.reshape(xx.shape)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.contour(xx, yy, Z, levels=20, cmap='Reds')
plt.scatter(combined_data['CPI Inflation'], combined_data['Market Return'],
            c=fin_labels, cmap='viridis', alpha=0.5, s=15)
plt.xlabel('CPI Inflation')
plt.ylabel('Market Return')
plt.title('GMM Density Contours (Negative Log Likelihood)')
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)
for i in range(3):
    # Plot each component's mean and covariance ellipse
    mean = gmm_fin.means_[i]
    cov = gmm_fin.covariances_[i]
    # For 2D, we can plot the 1-sigma ellipse
    # We'll just use a scatter of the component means
    plt.scatter(mean[0], mean[1], s=200, marker='X', label=f'Component {i+1}')
plt.scatter(combined_data['CPI Inflation'], combined_data['Market Return'],
            c=fin_labels, cmap='viridis', alpha=0.3, s=10)
plt.xlabel('CPI Inflation')
plt.ylabel('Market Return')
plt.title('Component Means')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('plots/gmm_contours.png', dpi=300)
plt.show()
```

**Key Insights:**
- **GMM provides soft assignments,** which are more informative than hard assignments when cluster boundaries are uncertain. The soft assignment visualisation shows how a point can have significant probability of belonging to multiple clusters.
- **AIC and BIC** are effective tools for selecting the number of components. AIC correctly identified 4 components (the true number), while BIC selected 3, favouring a simpler model. Both are valid, but AIC is less conservative and better at detecting the true number of components when the data is well‑separated.
- **In the financial case study,** GMM identified three market regimes: stable moderate returns, market downturns, and bull markets. The clusters were separated primarily along the return axis, suggesting that inflation had a relatively small impact on market returns over this period (the synthetic CPI data had low variance, while real CPI data may show different patterns).
- **GMM is more flexible than K‑Means** because it can model clusters with different shapes, sizes, and orientations (via covariance matrices). This makes it suitable for data with complex, overlapping cluster structures.
- The **density contours** visualisation shows the fitted Gaussian components, revealing that the bull market regime (cluster 2) had the widest spread in returns, while the downturn regime (cluster 1) was more tightly clustered around negative returns.
