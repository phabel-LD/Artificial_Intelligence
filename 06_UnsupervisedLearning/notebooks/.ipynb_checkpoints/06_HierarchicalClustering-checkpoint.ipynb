## 1. Theoretical Background

Hierarchical clustering builds a tree‑like structure (dendrogram) of clusters. There are two main approaches:

- **Agglomerative (Bottom‑Up):** Start with each point as its own cluster and iteratively merge the closest pairs.
- **Divisive (Top‑Down):** Start with all points in one cluster and recursively split.

**Key Concepts:**
- **Linkage Criteria:** How distance between clusters is computed:
  - **Single Linkage:** Minimum distance between any two points in the clusters.
  - **Complete Linkage:** Maximum distance between any two points.
  - **Average Linkage:** Average distance between all pairs.
  - **Ward Linkage:** Minimises the variance of the clusters being merged.
- **Dendrogram:** A tree diagram that shows the hierarchical relationship between clusters.
- **Cutting the Dendrogram:** Choosing a height at which to "cut" the tree to obtain a flat clustering.

## 2. Data Loading: Mall Customers Dataset

We load the Mall Customers dataset, which contains 200 customers with features: Gender, Age, Annual Income, and Spending Score. The goal is to segment customers into meaningful groups.


```python
# %% Data Loading – Mall Customers Dataset
import kagglehub
import pandas as pd
import numpy as np
from pathlib import Path

print("Loading Mall Customers dataset from Kaggle...")

try:
    # Use the correct dataset identifier
    path = kagglehub.dataset_download(
        "nalisha/mall-customer-segmentation-dataset-for-behavioral"
    )
    file_path = Path(path) / "Mall_Customers.csv"
    mc_df = pd.read_csv(file_path)
    print(f"Data loaded successfully from Kaggle.")
except Exception as e:
    print(f"Kaggle load failed: {e}")
    # Fallback: check local file
    local_path = Path("Mall_Customers.csv")
    if local_path.exists():
        mc_df = pd.read_csv(local_path)
        print(f"Loaded from local file: {local_path}")
    else:
        # Generate synthetic data (fallback)
        print("No external dataset found. Generating synthetic Mall Customers data...")
        np.random.seed(42)
        n_customers = 200
        age = np.random.normal(40, 15, n_customers).astype(int).clip(18, 70)
        income = np.random.normal(60, 30, n_customers).astype(int).clip(15, 140)
        spending_score = (100 - 0.5 * abs(income - 60) + np.random.normal(0, 15, n_customers)).clip(1, 100).astype(int)
        gender = np.random.choice(["Male", "Female"], n_customers, p=[0.4, 0.6])
        mc_df = pd.DataFrame({
            'CustomerID': range(1, n_customers + 1),
            'Gender': gender,
            'Age': age,
            'Annual Income (k$)': income,
            'Spending Score (1-100)': spending_score
        })
        print(f"Generated synthetic Mall Customers data with {n_customers} samples.")

# Check the dataset
print(f"\nDataset shape: {mc_df.shape}")
print("\nFirst 5 rows:")
print(mc_df.head())
print("\nSummary statistics:")
print(mc_df.describe())
```

```python
# %% Data Preprocessing
# Encode Gender (Male = 1, Female = 0)
mc_df["Gender"] = mc_df["Gender"].apply(lambda x: 1 if x == "Male" else 0)

# Select features for clustering (excluding CustomerID)
X = mc_df.iloc[:, 1:]  # Gender, Age, Annual Income, Spending Score

print(f"Feature matrix shape: {X.shape}")
print("\nFeature matrix (first 5 rows):")
print(X.head())
```

## 3. Dendrogram Visualisation

We generate a dendrogram using the Ward linkage method and Euclidean distance to understand the hierarchical structure of the data.

```python
# %% Dendrogram
# Compute linkage matrix
Z = linkage(X, method='ward', metric='euclidean')

# Plot dendrogram
plt.figure(figsize=(20, 15))
dendrogram(Z, leaf_rotation=90, leaf_font_size=8)
plt.title('Hierarchical Clustering Dendrogram (Ward Linkage)')
plt.xlabel('Customer Index')
plt.ylabel('Distance')
plt.axhline(y=200, color='red', linestyle='--', label='Cut at distance = 200')
plt.legend()
plt.tight_layout()
plt.savefig('plots/dendrogram.png', dpi=300)
plt.show()

print("The dendrogram shows the hierarchical structure of customer similarities.")
print("The red dashed line represents a cut at distance 200, suggesting ~5 clusters.")
```

## 4. Agglomerative Clustering

We apply agglomerative clustering with 5 clusters (based on the dendrogram) and visualise the results.

```python
# %% Agglomerative Clustering
n_clusters = 5
agg_clustering = AgglomerativeClustering(
    n_clusters=n_clusters,
    linkage='ward',
    metric='euclidean'
)
labels = agg_clustering.fit_predict(X)

# Add cluster labels to DataFrame
mc_df['cluster'] = labels

print(f"Cluster distribution:\n{mc_df['cluster'].value_counts().sort_index()}")

# Visualise clusters (Annual Income vs Spending Score)
plt.figure(figsize=(10, 8))
sns.scatterplot(
    x="Annual Income (k$)",
    y="Spending Score (1-100)",
    hue="cluster",
    palette="viridis",
    data=mc_df,
    s=60,
    alpha=0.7
)
plt.title('Customer Segments (Agglomerative Clustering, K=5)')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend(title='Cluster')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('plots/agglomerative_clusters.png', dpi=300)
plt.show()
```

### 4.1 Cluster Interpretation

We interpret each cluster based on its characteristics (income, spending score, age, gender).

```python
# %% Cluster Profiling
# Cluster summary statistics
cluster_summary = mc_df.groupby('cluster').agg({
    'Annual Income (k$)': ['mean', 'min', 'max'],
    'Spending Score (1-100)': ['mean', 'min', 'max'],
    'Age': 'mean',
    'Gender': lambda x: 'Male' if x.mean() > 0.5 else 'Female'
}).round(2)

print("Cluster Summary:")
print(cluster_summary)

# Visualise cluster profiles
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Age by cluster
age_summary = mc_df.groupby('cluster')['Age'].mean().reset_index()
sns.barplot(x='cluster', y='Age', data=age_summary, palette='viridis', ax=axes[0])
axes[0].set_title('Average Age by Cluster')
axes[0].set_xlabel('Cluster')
axes[0].set_ylabel('Average Age')

# Gender distribution by cluster
gender_counts = mc_df.groupby(['cluster', 'Gender']).size().reset_index(name='count')
gender_counts['Gender'] = gender_counts['Gender'].map({0: 'Female', 1: 'Male'})
sns.barplot(x='cluster', y='count', hue='Gender', data=gender_counts, palette='viridis', ax=axes[1])
axes[1].set_title('Gender Distribution by Cluster')
axes[1].set_xlabel('Cluster')
axes[1].set_ylabel('Count')

# Annual Income vs Spending Score by cluster
sns.scatterplot(
    x='Annual Income (k$)',
    y='Spending Score (1-100)',
    hue='cluster',
    palette='viridis',
    data=mc_df,
    s=40,
    alpha=0.6,
    ax=axes[2]
)
axes[2].set_title('Income vs Spending Score by Cluster')
axes[2].set_xlabel('Annual Income (k$)')
axes[2].set_ylabel('Spending Score (1-100)')

plt.tight_layout()
plt.savefig('plots/cluster_profiles.png', dpi=300)
plt.show()

print("\n=== Cluster Interpretations ===")
print("Cluster 0: Moderate income, moderate spending (middle segment)")
print("Cluster 1: High income, low spending (conservative customers)")
print("Cluster 2: High income, high spending (premium customers - most valuable)")
print("Cluster 3: Low income, high spending (spenders with budget constraints)")
print("Cluster 4: Low income, low spending (least engaged customers)")
```

## 5. Comparison with K‑Means

We compare hierarchical clustering results with K‑Means to validate the cluster structure.

```python
# %% K-Means Comparison
# Determine optimal K for K-Means using Elbow and Silhouette
k_range = range(2, 11)
inertia = []
silhouette_scores = []

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)
    sil_score = silhouette_score(X, kmeans.labels_)
    silhouette_scores.append(sil_score)

# Plot Elbow and Silhouette
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].plot(k_range, inertia, marker='o', linewidth=2)
axes[0].set_title('Elbow Method – K-Means')
axes[0].set_xlabel('Number of Clusters (k)')
axes[0].set_ylabel('Inertia')
axes[0].grid(True, alpha=0.3)

axes[1].plot(k_range, silhouette_scores, marker='o', linewidth=2, color='green')
axes[1].set_title('Silhouette Score – K-Means')
axes[1].set_xlabel('Number of Clusters (k)')
axes[1].set_ylabel('Silhouette Score')
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('plots/kmeans_optimal_k.png', dpi=300)
plt.show()

# Run K-Means with optimal K (based on silhouette)
best_k = k_range[np.argmax(silhouette_scores)]
print(f"Optimal K for K-Means (Silhouette): {best_k}")

kmeans = KMeans(n_clusters=best_k, random_state=42, n_init=10)
kmeans_labels = kmeans.fit_predict(X)
mc_df['cluster_kmeans'] = kmeans_labels

# Visualise K-Means clusters
plt.figure(figsize=(10, 8))
sns.scatterplot(
    x="Annual Income (k$)",
    y="Spending Score (1-100)",
    hue="cluster_kmeans",
    palette="viridis",
    data=mc_df,
    s=60,
    alpha=0.7
)
plt.title(f'Customer Segments (K-Means, K={best_k})')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend(title='Cluster')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('plots/kmeans_clusters.png', dpi=300)
plt.show()
```

### 5.1 Comparing Hierarchical vs K‑Means

We visualise the two clustering results side‑by‑side to compare the segmentations.

```python
# %% Side-by-Side Comparison
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Hierarchical clustering
sns.scatterplot(
    x="Annual Income (k$)",
    y="Spending Score (1-100)",
    hue="cluster",
    palette="viridis",
    data=mc_df,
    s=50,
    alpha=0.7,
    ax=axes[0]
)
axes[0].set_title('Hierarchical Clustering (Agglomerative)')
axes[0].set_xlabel('Annual Income (k$)')
axes[0].set_ylabel('Spending Score (1-100)')
axes[0].legend(title='Cluster')

# K-Means clustering
sns.scatterplot(
    x="Annual Income (k$)",
    y="Spending Score (1-100)",
    hue="cluster_kmeans",
    palette="viridis",
    data=mc_df,
    s=50,
    alpha=0.7,
    ax=axes[1]
)
axes[1].set_title(f'K-Means Clustering (K={best_k})')
axes[1].set_xlabel('Annual Income (k$)')
axes[1].set_ylabel('Spending Score (1-100)')
axes[1].legend(title='Cluster')

plt.tight_layout()
plt.savefig('plots/clustering_comparison.png', dpi=300)
plt.show()

# Silhouette score comparison
sil_hier = silhouette_score(X, mc_df['cluster'])
sil_kmeans = silhouette_score(X, mc_df['cluster_kmeans'])
print(f"Silhouette Score (Hierarchical): {sil_hier:.4f}")
print(f"Silhouette Score (K-Means):     {sil_kmeans:.4f}")
```
