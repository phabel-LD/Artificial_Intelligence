## 1. Data Loading and Preprocessing

We load the Netflix User Dataset from Kaggle and perform initial preprocessing.

```python
# %% Data Loading
print("Loading Netflix User Dataset from Kaggle...")
try:
    data = kagglehub.dataset_load(
        KaggleDatasetAdapter.PANDAS,
        "shivangi2018/netflix-user-dataset",
        "Netflix Userbase (1).csv",
    )
    print("Data loaded successfully from Kaggle.")
except Exception as e:
    print(f"Kaggle load failed: {e}")
    # Fallback: check local file
    local_path = "Netflix Userbase (1).csv"
    if Path(local_path).exists():
        data = pd.read_csv(local_path)
        print(f"Loaded from local file: {local_path}")
    else:
        raise FileNotFoundError("Dataset not found. Please download from Kaggle or provide a local copy.")

print(f"Dataset shape: {data.shape}")
print("\nFirst 5 rows:")
print(data.head())
print("\nSummary statistics:")
print(data.describe())
print("\nData types:")
print(data.dtypes)
```

```python
# %% Feature Engineering
# Drop 'User ID' (irrelevant for clustering)
data = data.drop('User ID', axis=1)

# Convert date columns to datetime
data['Join Date'] = pd.to_datetime(data['Join Date'], errors='coerce')
data['Last Payment Date'] = pd.to_datetime(data['Last Payment Date'], errors='coerce')

# Calculate active days
data['Active Days'] = (data['Last Payment Date'] - data['Join Date']).dt.days

# Drop date columns (no longer needed)
data = data.drop(['Last Payment Date'], axis=1)

# Remove rows with negative active days
data = data[data['Active Days'] >= 0]

print(f"Dataset shape after preprocessing: {data.shape}")
print("\nFirst 5 rows:")
print(data.head())
print("\nSummary statistics:")
print(data.describe(include='all'))
```

## 4. Categorical Clustering: K‑Modes

Since most variables are categorical, we use K‑Modes clustering for segmentation.

```python
# %% K-Modes Clustering
# Prepare categorical data (drop numerical vars, keep binned versions)
data['Age_Bin'] = pd.cut(data['Age'], bins=range(20, 60, 10))
data['Active_Days_Bin'] = pd.cut(data['Active Days'], bins=range(0, 800, 100))

data_kmodes = data.drop(['Age', 'Active Days'], axis=1)
data_kmodes = data_kmodes.drop(['Join Date Month', 'Join Month Label'], axis=1, errors='ignore')

# Drop any rows with missing values (safety check)
data_kmodes = data_kmodes.dropna()

# Encode categorical variables
encoder = OrdinalEncoder()
data_encoded = encoder.fit_transform(data_kmodes)
data_encoded = pd.DataFrame(data_encoded, columns=data_kmodes.columns)

# Find optimal number of clusters
silhouette_scores = []
k_range = range(2, 20)
for k in k_range:
    km = KModes(n_clusters=k, init='Huang', verbose=0, n_init=5, random_state=42)
    labels = km.fit_predict(data_encoded)
    score = silhouette_score(data_encoded, labels)
    silhouette_scores.append(score)

plt.figure(figsize=(10, 6))
plt.plot(k_range, silhouette_scores, marker='o')
plt.title('K‑Modes Silhouette Score vs. Number of Clusters')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.xticks(k_range)
plt.grid(True)
plt.tight_layout()
plt.savefig('plots/kmodes_silhouette.png', dpi=300)
plt.show()

best_k = k_range[np.argmax(silhouette_scores)]
print(f"Optimal k: {best_k}")

# Fit K-Modes with optimal k
km = KModes(n_clusters=best_k, init='Huang', n_init=10, verbose=1, random_state=42)
clusters = km.fit_predict(data_encoded)
data_kmodes['Cluster'] = clusters

# Cluster profiles
cluster_profiles = []
for cluster in sorted(data_kmodes['Cluster'].unique()):
    cluster_data = data_kmodes[data_kmodes['Cluster'] == cluster]
    profile = {'Cluster': cluster, 'Size': len(cluster_data)}
    for col in data_kmodes.columns[:-1]:
        profile[col] = cluster_data[col].mode()[0]
    cluster_profiles.append(profile)

profile_df = pd.DataFrame(cluster_profiles)
print("\nCluster Profiles (Mode values):")
print(profile_df.to_string(index=False))
```

```python
# %% K-Modes Visualisation
# Merge cluster labels from data_kmodes (which has the same index as the rows used)
data['Cluster'] = np.nan
data.loc[data_kmodes.index, 'Cluster'] = data_kmodes['Cluster']

# Drop rows that were not assigned a cluster (due to dropna)
data_vis = data.dropna(subset=['Cluster']).copy()

# Encode categorical columns for visualisation
encoder_vis = LabelEncoder()
cat_cols = ['Subscription Type', 'Country', 'Gender', 'Device']
for col in cat_cols:
    data_vis[col] = encoder_vis.fit_transform(data_vis[col])

fig = px.scatter_3d(data_vis, x='Age', y='Active Days', z='Monthly Revenue',
                    color=data_vis['Cluster'].astype(str),
                    title='K‑Modes Clustering (3D Projection)')
fig.update_layout(title_x=0.5, width=700, height=600)
fig.show()
```

## 5. Comparison with Other Clustering Methods

We compare K‑Modes results with numerical clustering methods on the scaled numerical features.

```python
# %% K-Means Clustering
from sklearn.cluster import KMeans

# Use numerical features
X_num = data[['Monthly Revenue', 'Age', 'Active Days']]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_num)

# Find optimal k
inertia = []
sil_scores = []
k_range = range(2, 15)
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)
    sil_scores.append(silhouette_score(X_scaled, kmeans.labels_))

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
axes[0].plot(k_range, inertia, marker='o')
axes[0].set_title('Elbow Method – K-Means')
axes[0].set_xlabel('k')
axes[0].set_ylabel('Inertia')
axes[0].grid(True)

axes[1].plot(k_range, sil_scores, marker='o', color='green')
axes[1].set_title('Silhouette Score – K-Means')
axes[1].set_xlabel('k')
axes[1].set_ylabel('Silhouette Score')
axes[1].grid(True)
plt.tight_layout()
plt.savefig('plots/kmeans_optimal_k.png', dpi=300)
plt.show()

best_k_kmeans = k_range[np.argmax(sil_scores)]
print(f"Optimal k for K‑Means: {best_k_kmeans}")

# Fit K-Means
kmeans_final = KMeans(n_clusters=best_k_kmeans, random_state=42, n_init=10)
kmeans_labels = kmeans_final.fit_predict(X_scaled)
data['KMeans_Cluster'] = kmeans_labels
print(f"K-Means silhouette: {silhouette_score(X_scaled, kmeans_labels):.4f}")
```

```python
# %% Hierarchical Clustering
agg = AgglomerativeClustering(n_clusters=3, linkage='ward')
hier_labels = agg.fit_predict(X_scaled)
data['Hier_Cluster'] = hier_labels
print(f"Hierarchical silhouette: {silhouette_score(X_scaled, hier_labels):.4f}")
```

```python
# %% DBSCAN and Spectral Clustering
# DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
db_labels = dbscan.fit_predict(X_scaled)
data['DBSCAN_Cluster'] = db_labels
print(f"DBSCAN clusters: {len(set(db_labels)) - (1 if -1 in db_labels else 0)}")

# Spectral Clustering
spectral = SpectralClustering(n_clusters=3, affinity='nearest_neighbors', random_state=42)
spec_labels = spectral.fit_predict(X_scaled)
data['Spectral_Cluster'] = spec_labels
print(f"Spectral silhouette: {silhouette_score(X_scaled, spec_labels):.4f}")
```

```python
# %% GMM
gmm = GaussianMixture(n_components=3, random_state=42)
gmm_labels = gmm.fit_predict(X_scaled)
data['GMM_Cluster'] = gmm_labels
print(f"GMM silhouette: {silhouette_score(X_scaled, gmm_labels):.4f}")
```

## 6. Factor Analysis

We apply exploratory factor analysis to understand latent factors underlying the data.

```python
# %% Factor Analysis
# Use the encoded categorical data
fa_data = data_encoded.copy()

# 1. Drop any rows with missing values (safety)
fa_data = fa_data.dropna()

# 2. Remove constant columns (variance = 0)
constant_cols = [col for col in fa_data.columns if fa_data[col].std() == 0]
if constant_cols:
    print(f"Removing constant columns: {constant_cols}")
    fa_data = fa_data.drop(columns=constant_cols)

# 3. Check KMO and Bartlett (they will be NaN if correlation matrix is singular)
try:
    kmo_all, kmo_model = calculate_kmo(fa_data)
    print(f"KMO Score: {kmo_model:.4f}")
    print("Interpretation: " + ("Good" if kmo_model > 0.6 else "Poor"))
except Exception as e:
    print(f"KMO calculation failed: {e}")

try:
    chi2, p = calculate_bartlett_sphericity(fa_data)
    print(f"Bartlett's Test: χ² = {chi2:.2f}, p = {p:.4e}")
except Exception as e:
    print(f"Bartlett's test failed: {e}")

# 4. Fit Factor Analysis - use at most 3 factors or the number of features
n_factors = min(3, fa_data.shape[1])   # <-- replaced CONFIG lookup

fa = FactorAnalyzer(n_factors=n_factors, rotation='varimax')
try:
    fa.fit(fa_data)
    loadings = pd.DataFrame(fa.loadings_, index=fa_data.columns)
    print("\nFactor Loadings:")
    print(loadings.round(3))
    
    eigenvals = fa.get_eigenvalues()
    print(f"\nEigenvalues (first {len(eigenvals[0])}): {eigenvals[0].round(3)}")
    
    # Scree plot
    plt.figure(figsize=(8, 5))
    plt.plot(range(1, len(eigenvals[0])+1), eigenvals[0], marker='o')
    plt.axhline(y=1, color='red', linestyle='--', label='Kaiser Criterion')
    plt.title('Scree Plot – Factor Analysis')
    plt.xlabel('Factor Number')
    plt.ylabel('Eigenvalue')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.savefig('plots/factor_analysis_scree.png', dpi=300)
    plt.show()
    
except Exception as e:
    print(f"Factor analysis fitting failed: {e}")
    print("Trying a fallback using PCA-based factor extraction (no rotation)...")
    # Fallback to PCA
    from sklearn.decomposition import PCA
    pca = PCA(n_components=min(3, fa_data.shape[1]))
    pca.fit(fa_data)
    loadings = pd.DataFrame(pca.components_.T, index=fa_data.columns,
                            columns=[f"Factor{i+1}" for i in range(pca.n_components_)])
    print("\nPrincipal Component Loadings (used as fallback):")
    print(loadings.round(3))
    print(f"\nExplained variance ratio: {pca.explained_variance_ratio_.round(3)}")
    # Scree plot from PCA
    plt.figure(figsize=(8,5))
    plt.plot(range(1, len(pca.explained_variance_)+1), pca.explained_variance_, marker='o')
    plt.axhline(y=1, color='red', linestyle='--', label='Kaiser Criterion')
    plt.title('Scree Plot (PCA)')
    plt.xlabel('Principal Component')
    plt.ylabel('Eigenvalue')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.savefig('plots/factor_analysis_scree_fallback.png', dpi=300)
    plt.show()
```

**Key Insights:**
- **K‑Modes is the most suitable clustering method** for this dataset because it respects the categorical nature of most variables. The two clusters reveal distinct behavioural segments: Cluster 0 (higher revenue, longer active, Standard plan) and Cluster 1 (lower revenue, shorter active, Basic plan).
- **Numerical clustering methods are less effective** due to the weak linear structure and the dominance of categorical information – the low silhouette scores reflect that the numerical features alone do not form natural groupings.
- **Factor analysis has limited utility** given the low KMO and the dominance of a single variable (`Active Days_Bin`) in the first factor, making latent structure interpretation difficult.
- **Business Implications:** Cluster 0 represents “premium/engaged” users who could be targeted for upselling or loyalty programmes, while Cluster 1 represents “basic/price‑sensitive” users who may benefit from retention campaigns to extend their subscription duration.

**Suggested Next Steps:**
1. **Validate Cluster Stability:** Use bootstrapping or cross‑validation to confirm that the two K‑Modes clusters are robust across different random initialisations.
2. **Incorporate Additional Features:** Add behavioural data (e.g., viewing history, genre preferences, payment method) to refine segmentation and potentially uncover finer subgroups.
3. **Dynamic Segmentation:** Monitor cluster membership over time (e.g., quarterly) to detect shifts in customer behaviour and adjust marketing strategies accordingly.
4. **Experiment with Hybrid Methods:** Try **K‑Prototypes** to handle mixed numeric‑categorical data without binning, potentially yielding more nuanced clusters.
5. **Use Clusters for Predictive Modelling:** Include cluster labels as a feature in churn prediction or lifetime value models to evaluate their predictive power.
6. **Automate Clustering in Production:** Build a pipeline that assigns new users to the two clusters in real time for personalised campaigns.
7. **Improve Factor Analysis:** Collect more diverse variables (e.g., engagement metrics) to increase KMO and enable meaningful latent factor extraction, or use PCA directly as a dimensionality reduction step before clustering.
8. **Visualise Clusters with PCA on Encoded Data:** Apply PCA to the encoded categorical data and project the two clusters in 2D to visualise separation (even if the numerical features are weak).
