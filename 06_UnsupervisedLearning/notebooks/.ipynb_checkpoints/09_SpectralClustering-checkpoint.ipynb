## 1. Theoretical Background

Spectral Clustering is a graph‑based clustering method that uses the eigenvectors of a similarity (affinity) matrix to perform dimensionality reduction before applying a standard clustering algorithm (e.g., K‑Means).

**Key Concepts:**
- **Similarity Matrix:** A matrix where each entry represents the similarity between two points. Common kernels: RBF (Gaussian), k‑nearest neighbours.
- **Graph Laplacian:** A matrix derived from the similarity matrix. The eigenvectors of the Laplacian are used to embed the data into a lower‑dimensional space.
- **Spectral Embedding:** The projection of data onto the eigenvectors of the Laplacian.
- **Normalised vs Unnormalised Laplacian:** Normalised Laplacians (e.g., `'random_walk'`) often perform better on real‑world data.

**Advantages of Spectral Clustering:**
- Can find clusters of arbitrary shape.
- Does not assume spherical clusters.
- Works well on non‑linear data.

**Disadvantages:**
- Computationally expensive for large datasets.
- Sensitive to the choice of similarity kernel and parameters.

In this notebook, we apply spectral clustering to image segmentation tasks and the classic moons dataset.

## 2. Synthetic Image: Four Circles

We generate a synthetic image containing four circles with different centres and radii. Spectral clustering will segment the image into the four circles.

```python
# %% Generate Synthetic Image
l = 100
x, y = np.indices((l, l))

# Circle centres and radii
center1 = (28, 24)
center2 = (40, 50)
center3 = (67, 58)
center4 = (24, 70)

radius1, radius2, radius3, radius4 = 16, 14, 15, 14

# Create circles
circle1 = (x - center1[0]) ** 2 + (y - center1[1]) ** 2 < radius1 ** 2
circle2 = (x - center2[0]) ** 2 + (y - center2[1]) ** 2 < radius2 ** 2
circle3 = (x - center3[0]) ** 2 + (y - center3[1]) ** 2 < radius3 ** 2
circle4 = (x - center4[0]) ** 2 + (y - center4[1]) ** 2 < radius4 ** 2

# Combine circles
img = circle1 + circle2 + circle3 + circle4
mask = img.astype(bool)

# Add noise
img = img.astype(float)
img += 1 + 0.2 * np.random.randn(*img.shape)

print(f"Image shape: {img.shape}")
print(f"Number of pixels: {img.size}")
print(f"Number of non‑zero pixels: {mask.sum()}")

# Visualise the image
plt.figure(figsize=(8, 8))
plt.matshow(img, fignum=0, cmap='gray')
plt.title('Synthetic Image: Four Circles with Noise')
plt.axis('off')
plt.tight_layout()
plt.savefig('plots/circles_image.png', dpi=300)
plt.show()
```

### 2.1 Building the Similarity Graph

We convert the image to a graph using `image.img_to_graph`, where each pixel is a node and edges are weighted by pixel intensity similarity. The similarity is then kernelised using an exponential function.

```python
# %% Build Similarity Graph
# Convert image to graph
graph = image.img_to_graph(img, mask=mask)

# Kernelise the graph (RBF kernel)
graph.data = np.exp(-graph.data / graph.data.std())

print(f"Graph shape: {graph.shape}")
print(f"Number of non‑zero edges: {graph.nnz}")
```

### 2.2 Spectral Embedding + K‑Means

We first compute the spectral embedding (eigenvectors of the Laplacian) and then apply K‑Means on the embedded space.

```python
# %% Spectral Embedding + K-Means
embedding = SpectralEmbedding(n_components=4, affinity='precomputed', random_state=42)
graph_embedded = embedding.fit_transform(graph.toarray())

# Apply K-Means on the embedding
kmeans = KMeans(n_clusters=4, random_state=42)
labels = kmeans.fit_predict(graph_embedded)

# Reconstruct the labelled image
label_im = np.full(mask.shape, -1.0)
label_im[mask] = labels

# Visualise
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
axes[0].matshow(img, cmap='gray')
axes[0].set_title('Original Image')
axes[0].axis('off')

axes[1].matshow(label_im, cmap='viridis')
axes[1].set_title('Spectral Clustering (Embedding + K-Means)')
axes[1].axis('off')

plt.tight_layout()
plt.savefig('plots/spectral_embedding_kmeans.png', dpi=300)
plt.show()

print(f"Number of unique labels: {len(np.unique(labels))}")
```

### 2.3 Direct Spectral Clustering

Scikit‑learn provides a direct implementation of spectral clustering that combines the embedding and clustering steps.

```python
# %% Direct Spectral Clustering
from sklearn.cluster import spectral_clustering

labels_direct = spectral_clustering(graph, n_clusters=4, eigen_solver='arpack', random_state=42)

# Reconstruct the labelled image
label_im_direct = np.full(mask.shape, -1.0)
label_im_direct[mask] = labels_direct

# Visualise
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
axes[0].matshow(img, cmap='gray')
axes[0].set_title('Original Image')
axes[0].axis('off')

axes[1].matshow(label_im_direct, cmap='viridis')
axes[1].set_title('Direct Spectral Clustering')
axes[1].axis('off')

plt.tight_layout()
plt.savefig('plots/direct_spectral_clustering.png', dpi=300)
plt.show()

print(f"Number of unique labels: {len(np.unique(labels_direct))}")
```

## 3. Moons Dataset: Non‑Linear Shape

We apply spectral clustering to the classic moons dataset, which is non‑linearly separable. K‑Means fails on this dataset, but spectral clustering successfully separates the two moons.

```python
# %% Moons Dataset
X_moons, y_moons = make_moons(n_samples=200, noise=0.1, random_state=42)

# Apply spectral clustering
spectral = SpectralClustering(
    n_clusters=2,
    affinity='nearest_neighbors',
    n_neighbors=10,
    random_state=42,
    assign_labels='cluster_qr'
)
labels_spectral = spectral.fit_predict(X_moons)

# Visualise results
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Original data
axes[0].scatter(X_moons[:, 0], X_moons[:, 1], c=y_moons, cmap='coolwarm', alpha=0.7)
axes[0].set_title('Moons Dataset (True Labels)')
axes[0].set_xlabel('Feature 1')
axes[0].set_ylabel('Feature 2')
axes[0].grid(True, alpha=0.3)

# Spectral clustering
axes[1].scatter(X_moons[:, 0], X_moons[:, 1], c=labels_spectral, cmap='coolwarm', alpha=0.7)
axes[1].set_title('Spectral Clustering (Predicted)')
axes[1].set_xlabel('Feature 1')
axes[1].set_ylabel('Feature 2')
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('plots/moons_spectral_clustering.png', dpi=300)
plt.show()

# Compare with K-Means
from sklearn.cluster import KMeans
kmeans_moons = KMeans(n_clusters=2, random_state=42)
labels_kmeans = kmeans_moons.fit_predict(X_moons)

# Silhouette scores
from sklearn.metrics import silhouette_score
sil_spectral = silhouette_score(X_moons, labels_spectral)
sil_kmeans = silhouette_score(X_moons, labels_kmeans)

print(f"Silhouette Score (Spectral): {sil_spectral:.4f}")
print(f"Silhouette Score (K-Means):  {sil_kmeans:.4f}")
```

## 4. Comparison of Affinity Types

We explore different affinity (similarity) kernels for spectral clustering on the moons dataset.

```python
# %% Affinity Comparison
affinities = ['rbf', 'nearest_neighbors']
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

for idx, affinity in enumerate(affinities):
    # Build parameters conditionally
    kwargs = {
        'n_clusters': 2,
        'affinity': affinity,
        'random_state': 42,
        'assign_labels': 'cluster_qr'
    }
    if affinity == 'nearest_neighbors':
        kwargs['n_neighbors'] = 10
    
    spectral_aff = SpectralClustering(**kwargs)
    labels_aff = spectral_aff.fit_predict(X_moons)
    
    axes[idx].scatter(X_moons[:, 0], X_moons[:, 1], c=labels_aff, cmap='coolwarm', alpha=0.7)
    axes[idx].set_title(f'Affinity: {affinity}')
    axes[idx].set_xlabel('Feature 1')
    axes[idx].set_ylabel('Feature 2')
    axes[idx].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('plots/affinity_comparison.png', dpi=300)
plt.show()

print("Both affinity types perform well on the moons dataset.")
print("'nearest_neighbors' is often more robust to parameter choice.")
```
